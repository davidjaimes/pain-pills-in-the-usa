```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#pills data
pills = pd.read_csv('pills-per-zipcode.csv')

#race data
demo_hispanic = pd.read_csv('2010Census_DemoProfile3b_Hispanic.csv')
df2 = demo_hispanic[['Geography','Hispanic or Latino (of any race)']]
df2zip = df2.rename(columns={"Geography":"Zip Code","Hispanic or Latino (of any race)":"Hispanic"})

demo_bw = pd.read_csv('2010Census_White_AfricanAmerican.csv')
df3 = demo_bw[['Geography','White','Black or African American']]
df3zip = df3.rename(columns={"Geography":"Zip Code"})


#merge
merge_table = pd.merge(pills, df2zip, on="Zip Code")
merge_table
full_table = pd.merge(merge_table, df3zip, on="Zip Code")
full_table
```
