```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/merged.csv')
df
```

```python
plt.plot(df['Pills'], df['Death Rate'], 'o')
topten = df.sort_values(by=['Death Rate']).tail(10)
#print(topten)
plt.plot(topten['Pills'], topten['Death Rate'], 'o')

plt.xscale('log')
plt.yscale('log')


from scipy import stats
stats.ttest_1samp(topten['Pills'], df['Pills'].mean())
topten[topten['Pills'] > df['Pills'].mean()]
```

```python
pop_income = df['AGI thousands'] * 1e3 / df['Tax Returns']
plt.plot(pop_income, df['Pills'], 'o')
plt.xscale('log')
plt.yscale('log')

v = pop_income.mean()
```

```python
w1 = pop_income < 50000
w2 = (pop_income >= 50000) & (pop_income < 100000)
w3 = pop_income >= 100000
plt.axvline(v, color='k', linestyle='solid')
plt.plot(pop_income[w1], df['Pills'][w1], 'o', color='b', label='Low (<$50,000)')
plt.plot(pop_income[w2], df['Pills'][w2], 'o', label='Medium')
plt.plot(pop_income[w3], df['Pills'][w3], 'o', label='High (>=$100,000)')
plt.grid(True)
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Pills')
plt.xlabel('Income')
plt.legend(loc='upper right')
w1
```

```python
stats.f_oneway(df['Pills'][w1], df['Pills'][w2], df['Pills'][w3])
```

```python
df['Group'] = ''
```

```python
df['Group'][w1] = 'Low'
df['Group'][w2] = 'Medium'
df['Group'][w3] = 'High'
df
```

```python
df.boxplot('Pills', by='Group')
plt.ylim(1e5, 1e7)
```
