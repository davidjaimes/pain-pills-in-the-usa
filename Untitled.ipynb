```python
import pandas as pd
```

```python
path = 'David/census_2010_population_zcta5.csv'
pop = pd.read_csv(path)
pop
```

```python
path = 'David/zip_value_counts.csv'
valcnt = pd.read_csv(path)
valcnt
```

```python
merged = pd.merge(valcnt, pop, how='left', left_on='Zip Code', right_on='ZCTA5')
merged
```

```python
del merged['ZCTA5']
merged
```

```python
merged = merged.dropna()
merged
```

```python
merged = merged.astype('int32')
merged
```

```python
merged = merged.rename(columns={'Counts': 'Transactions', 'Total': '2010 Population'})
merged
```

```python
merged = merged.rename(columns={'Transactions': '2006-2012 Transactions'})
merged
```

```python
merged.to_csv('David/summary.csv', index=False)
```
