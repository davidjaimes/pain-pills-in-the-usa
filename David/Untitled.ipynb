```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pills = pd.read_csv('2010_pills.csv')
zipcode = pills.sort_values(by='2010 Pills', ascending=False).head(300)['Zip Code']
```

```python
income = pd.read_csv('CA_2010_income_pivot.csv', index_col=['Zip Code'])
income = income.fillna(0)
inrng = list(income.columns)
values = np.zeros(len(inrng))
for zc in zipcode:
    values += income[income.index == zc].values.flatten()
plt.barh(inrng, values / 1e6, zorder=3)
plt.title('Top 300 Zip Codes', size=16)
plt.xlabel('2010 Individual Tax Returns (million)', size=12)
plt.ylabel('Size of Adjusted Gross Income', size=12)
plt.xlim(0, 90)
plt.grid()
plt.tight_layout()
plt.savefig('income-top-300-zipcodes.png', dpi=150)
```

```python
income = pd.read_csv('CA_2010_income_pivot.csv', index_col=['Zip Code'])
inrng = list(income.columns)
values = income[income.index == 94550].values.flatten() / 1e3
plt.barh(inrng, values, zorder=3)
plt.title('Livermore, CA 94550', size=16)
plt.xlabel('2010 Individual Tax Returns (thousand)', size=12)
plt.ylabel('Size of Adjusted Gross Income', size=12)
plt.xlim(0, 800)
plt.grid()
plt.tight_layout()
plt.savefig('income-livermore.png', dpi=150)
```
